```python
import pandas as pd
import os
```

```python


data_path="/opt/airflow/data/raw/us_airline_sentiment_raw.csv"
data_processed_path="/opt/airflow/data/processed/us_airline_sentiment_processed.csv"
df=pd.read_csv(data_path)
df.head()

df["text_clean"] = (
    df["text"]
    .str.lower()                              
    .str.replace(r"[^\w\s]", "", regex=True) 
    .str.replace(r'http\S+|www\S+|https\S+', '', regex=True)
    .str.replace(r'#', '', regex=True)
    .str.replace(r'@\w+', '', regex=True)
    .str.replace('\n', ' ')  
    .str.strip()                            
)
print(df[df["text_clean"].duplicated(keep=False)]["text_clean"].value_counts())
text_counts = df["text_clean"].value_counts()
duplicates = text_counts[text_counts > 1]
print(duplicates)

df['text_clean']=df['text_clean'].str.replace('virginamerica','')
df['text_clean']=df['text_clean'].str.replace('united','')
df['text_clean']=df['text_clean'].str.replace('southwestair','')
df['text_clean']=df['text_clean'].str.replace('jetblue','')
df['text_clean']=df['text_clean'].str.replace('americanair','')
df['text_clean']=df['text_clean'].str.replace('usairways','')
df.head()
        



df_clean = df[['tweet_id', 'text_clean', 'airline_sentiment']].copy()
df_clean.rename(columns={'tweet_id': 'id', 'airline_sentiment': 'label'}, inplace=True)
final_df = df_clean[['id', 'label', 'text_clean']]
display(final_df.text_clean.tail())
```

```python

print("Nombre de doublons avant :", df["text_clean"].duplicated().sum())

df = df.drop_duplicates(subset="text_clean", keep="first")

print("Nombre de doublons après :", df["text_clean"].duplicated().sum())


final_df.to_csv(data_processed_path,index=False)
```
